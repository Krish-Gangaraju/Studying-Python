```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('/Users/kirangangaraju/Desktop/Python/ZugZwang/ML/CSVs/trial.csv')
df
```

### 1. Countplot() --> count instance of each category

```python
sn.countplot(df['COUNTRY'])
```

```python
fig, ax = plt.subplots(figsize=(10,5))            # Changing size of graph
sn.countplot(df['COUNTRY'], ax=ax, data=df)

plt.title("Players in the IPL", fontsize = 22)    # Adding titles
plt.xlabel("Country", fontsize = 18)              
plt.ylabel("Total Players", fontsize = 18)

plt.savefig("Country vs Players")
```

```python
fig, ax = plt.subplots(figsize=(10,5))   
sn.countplot(df['PLAYING ROLE'], ax=ax, data=df)

plt.title("Player Style", fontsize = 22)  
plt.xlabel("Style", fontsize = 18)              
plt.ylabel("No. of Players", fontsize = 18)
```

```python
fig, ax = plt.subplots(figsize=(10,5))   
sn.countplot(df['TEAM'], ax=ax, data=df)

plt.title("IPL Teams", fontsize = 22)  
plt.xlabel("Team", fontsize = 18)              
plt.ylabel("No. of Players", fontsize = 18)
```

```python
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header = None)
columns = ["Sepal-l", "Sepal-w","Petal-l","Petal-w","Flower"]
iris.columns = columns

iris['Flower'].value_counts()
```

```python
iris['Flower'].value_counts().plot.pie()
plt.show()
```

```python
iris['Flower'].value_counts().plot.pie(explode = [0.1 , 0.1 , 0.1], autopct = '%1.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12,10))
df['COUNTRY'].value_counts().plot.pie(autopct = '%1.1f%%', ax = ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,5))
df['PLAYING ROLE'].value_counts().plot.pie(autopct = '%1.1f%%', ax = ax)
plt.show()
```

### Barplot() - Bar Graph

```python
sixers = df[['PLAYER NAME' , 'SIXERS']]
sixers.head()
```

```python
sorted_sixers = sixers.sort_values(by = "SIXERS", ascending = False)
```

```python
top20 = sorted_sixers[:20]
top20.head(20)
```

```python
fig, ax = plt.subplots(figsize = (12,6))
sn.barplot(top20['PLAYER NAME'], top20['SIXERS'], ax=ax, data = top20)

plt.title("Top 20 Six Hitters", fontsize = 18)
plt.xlabel("Batsman", fontsize = 16)
plt.ylabel("Sixers", fontsize = 16)
plt.xticks(rotation = 80)

plt.savefig("Hitters.jpg")
```

### Multiple Barplot() - Bar Graph

```python
select = df[['COUNTRY' , "PLAYING ROLE" , "SOLD PRICE"]]
select.head()
```

```python
ipltrial = select.groupby(['COUNTRY' , 'PLAYING ROLE']).mean().round(3)
ipltrial.head()
```

```python
ipltrial.reset_index().head(10)
```

```python
plt.figure(figsize = (10,8))
sn.barplot(x = 'COUNTRY' , y = "SOLD PRICE" , hue = "PLAYING ROLE" , data=df , ci=None)

plt.title("Price Distribution", fontsize = 18)
plt.xlabel("COUNTRY", fontsize = 16)
plt.ylabel("SOLD PRICE", fontsize = 16)
```
