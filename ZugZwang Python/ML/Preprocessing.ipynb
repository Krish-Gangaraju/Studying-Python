### Groupby

```python
batsman = df[df["PLAYING ROLE"]=='Batsman'][['COUNTRY',"T-RUNS","ODI-RUNS-S"]]
batsman.head()
```

```python
batsman_mean = batsman.groupby(by = ["COUNTRY"]).mean()
batsman_mean
```

```python
batsman_mean = batsman.groupby(by = ["COUNTRY"]).max()
batsman_mean
```

```python
batsman_mean = batsman.groupby(by = ["COUNTRY"]).median()
batsman_mean
```
