## Loc - Indexer

```python
df.loc[[1,10,100]]   # Get only Row 1,10,100 with all Columns
```

```python
df.loc[[1,10,100]][["PLAYER NAME","COUNTRY"]]     # Row 1,10,100 with only these Columns
```

```python
df.loc[2:5,["PLAYER NAME","COUNTRY"]]      # Includes 5 and uses Location
```

```python
df[2:5][["PLAYER NAME","COUNTRY"]]    # Doesn't include 5 and doesn't use Location
```

```python
trial = df.set_index("PLAYER NAME")   # Making PLAYER NAME as Index
trial.head()
```

```python
# Now observe that trial[2:5] will work but trial.loc[2:5] will not

# trial [2:5]             # - Index number will still work
# trial.loc[2:5]          # - Index number won't work as there is no label from 2 to 5 anymore
trial.loc["Abdulla, YA" : "Badrinath, S"]      # - Player Name is new Index so it will work
```

```python
# Parameter 1 for rows and Parameter 2 for columns
df.loc[[0,7],:]      # Only 0 and 7 with all Columns
```

```python
df.loc[0:7, :]       # From 0 to 7 with all Columns
```

```python
df.loc[0:3, ['PLAYER NAME', 'COUNTRY', 'TEAM']]     # From 0 to 3 with only those 3 Columns
```

## ILoc - Indexer where I stands for Integer

```python
df.iloc[5:11 , 2:7]    # Using integers instead of numbers. Works like range doesn't include last
```

```python
df.iloc[5:11 , ]     # Only 5-10 Rows but all Columns
```

```python
df.iloc[:, 5:11 ]    # Shows all Rows but the COLON is NECESSARY
```

```python
# You can do other stuff like

# df.iloc [:,:]
# df.iloc [:5 , :5]
# etc.
```

## Subetting with Criteria (ascending/descending)

```python
df.sort_values (by=['T-RUNS'],ascending=False)[0:5]    # Descending according to most Test Runs
```

```python
df.sort_values (by = ["SIXERS"] , ascending = False)[0:1][["PLAYER NAME"]]
```

```python
df.sort_values (by = ["HS"] , ascending = False).iloc[0]["PLAYER NAME"]
```

```python
df.sort_values (by = ["SOLD PRICE"] , ascending = False).iloc[0]["PLAYER NAME"]
```
