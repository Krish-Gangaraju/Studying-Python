```python
import pandas as pd
```

### Apply and Sum

```python
data_m = {
    "Names":("John","David","Chris") , "Math":(89,90,91) , "Phy":(92,92,94) , "Chem":(69,88,93)
         }
df_m = pd.DataFrame(data_m)
df_m
```

```python
# i) with separate function
df_total_func = df_m.copy()
def total(row):
    return row["Math"] + row["Phy"] + row["Chem"]
df_total_func["Total"] = df_total_func.apply(total, axis=1)
df_total_func
```

```python
# ii) with lambda function
df_total_lamb = df_m.copy()
df_total_lamb["Total"] = df_total_lamb.apply(lambda row : row["Math"] + row["Phy"] + row["Chem"] , axis=1)
df_total_lamb
```

```python
import numpy as np
```

```python
# iii) with numpy after removing the "Name" and making that the indexing
df_total_np = df_m.copy()
df_total_np.set_index("Names", inplace=True)
df_total_np["Total"] = df_total_np.apply(np.sum , axis = 1)
df_total_np
```

```python
# iv) with inbuilt apply command 'sum'
df_total = df_m.copy()
df_total["Total"] = df_total.apply('sum' , axis = 1)
df_total
```

```python
df_m
```

```python
# v) .agg
df_total = df_m.copy()
df_total["Total"] = df_total.agg('sum',axis=1)
df_total
```

```python
# vi) df.sum
df_total_sum = df_m.copy()
df_total_sum["Total"] = df_total_sum.sum(axis=1)
df_total_sum
```

```python
df_nan = pd.DataFrame( [[89,np.nan,69] , [90,92,88] , [91,94,93]] , columns = ["Math","Phy","Chem"])
def total(row):
    row = row.fillna(0)
    return row["Math"] + row["Phy"] + row["Chem"]
df_nan["Total"] = df_nan.apply(total, axis=1)
df_nan
```
