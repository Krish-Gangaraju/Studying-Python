#### Getting data from Multindexing table

```python
venue_index = pd.MultiIndex.from_product([["India","Aus"] ,["Batsman","Bowler"]], names=["Country","Role"])
venue_columns = pd.MultiIndex.from_product([["2018","2019","2020"] , ["InIND","InAUS"]], names=["Year","Venue"])
numbers = np.random.randint(20,50, size=(4,6))
venues = pd.DataFrame(numbers, index=venue_index, columns=venue_columns)
venues
```

```python
venues[[('2019','InIND')]] # doesn't remove the names/titles
```

```python
venues[[('2019','InIND') , ('2020','InAUS')]]
```

```python
venues.iloc[[3]]         # only row
```

```python
venues.iloc[[3]][[('2019','InIND')]]     # row and columns
```

```python
venues.loc[("India","Batsman")][("2020","InIND")]     # not a dataframe
```

```python
venues.loc[('Aus','Batsman')][("2018","InIND")]
```

```python
venues.loc[('India','Bowler')]
```
