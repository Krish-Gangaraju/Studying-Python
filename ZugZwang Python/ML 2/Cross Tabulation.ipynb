```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.columns = ["Total","Tip","Gender","Smoker","Day","Time","Guests"]
tips.head()
```

```python
# index = columns that becomes the index
# columns = columns whose categories becomes individual columns
# aggfunc = aggregate function that is applied
# values = column that aggfunc is applied on
```

```python
pd.crosstab(index=tips['Day'], columns=tips['Gender'])
```

```python
tips_1 = pd.crosstab(index=tips['Day'], columns=tips['Gender'], values=tips['Total'], aggfunc='sum')
tips_1
```

```python
tips_2 = pd.crosstab(index=tips['Day'], columns=tips['Gender'], values=tips['Guests'], aggfunc='sum')
tips_2
```

```python
pd.crosstab(index=tips['Day'], columns=tips['Gender'],values=tips['Guests'],aggfunc='sum').rename(columns={"Male": "Guests_served_by_males", "Female": "Guests_served_by_females"})
```

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['infantry', 'infantry', 'cavalry', 'cavalry', 'infantry', 'infantry', 'cavalry', 'cavalry','infantry', 'infantry', 'cavalry', 'cavalry'], 
        'experience': ['veteran', 'rookie', 'veteran', 'rookie', 'veteran', 'rookie', 'veteran', 'rookie','veteran', 'rookie', 'veteran', 'rookie'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
army = pd.DataFrame(raw_data, columns = ['regiment', 'company', 'experience', 'name', 'preTestScore', 'postTestScore'])
army
```
